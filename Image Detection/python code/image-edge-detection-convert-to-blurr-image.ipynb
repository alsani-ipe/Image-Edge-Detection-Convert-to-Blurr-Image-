# **Edge Detection by using Sobel Operator**

```python
argumentimage= { "image": " ../input/castimage/cast.jpg","scharr": 0}
```

```python
bgrimage=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#1st step: Convert RGB Image into Gray Image

grayimage= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(bgrimage)

plt.show()
```

```python
plt.imshow(grayimage)
```

```python
#2nd step: make a kernel with scharr=0 value

kernelGen= -1 if argumentimage["scharr"]> 0 else 3
```

```python
#3rd step: implementation of Sobel Operator by OpenCV

gradientImageX= cv2.Sobel(grayimage, ddepth=cv2.CV_32F, dx=1,dy=0, ksize=kernelGen)
gradientImageY= cv2.Sobel(grayimage, ddepth=cv2.CV_32F, dx=0,dy=1, ksize=kernelGen)
```

```python
castimageShow("Gradient Image",gradientImageX)
castimageShow("Gradient Image",gradientImageY)
```

# **Convert Orginal Image into Blur Image By using GaussianBlur Operator**

```python
image=cv2.imread("../input/castimage/cast.jpg")
grayimage= cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurimage= cv2.GaussianBlur(grayimage,(5,5),0)
castimageShow("Orginal Image",image)
castimageShow("Blur Image",blurimage)
```

# **Edge Detection By using Canny Operator**

```python
image=cv2.imread("../input/castimage/cast.jpg")
```

```python
cannyBlurWide=cv2.Canny(blurimage,20,250)
cannyBlurMiddle=cv2.Canny(blurimage,240,250)
cannyBlurFit=cv2.Canny(blurimage,10,255)

castimageShow("Orginal Image",image)
castimageShow("cannyBlurWide",cannyBlurWide)
castimageShow("cannyBlurMiddle",cannyBlurMiddle)
castimageShow("cannyBlurFit",cannyBlurFit)
```
